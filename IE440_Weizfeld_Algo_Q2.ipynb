```python
import pandas as pd
import numpy as np
```

```python
D=pd.read_csv('g 4_demands.csv',header=None)
C=pd.read_csv('g 4_costs.csv',header=None)
Cord=pd.read_csv('g 4_coordinates.csv',header=None)
```

```python
def euclidean_dist(x1,x2,a1,a2):
    return(np.sqrt((x1-a1)**2+(x2-a2)**2))
    
```

```python
def Weiszfeld_Algo():
    
    x1_=[0,np.sum(C.loc[:,0]*D.loc[:,0]*Cord.loc[:,0])/\
            np.sum(C.loc[:,0]*D.loc[:,0])]   ### initial x1 value for the single facility
    x2_=[0,np.sum(C.loc[:,0]*D.loc[:,0]*Cord.loc[:,1])/\
            np.sum(C.loc[:,0]*D.loc[:,0])]   ### initial x2 value for the single facility
    
    while euclidean_dist(x1_[-1],x2_[-1],x1_[-2],x2_[-2])>=1e-6: ### stop when euclidean distance between iterations is small enough
        
        x1_new=np.sum(C.loc[:,0]*D.loc[:,0]*Cord.loc[:,0]/euclidean_dist(x1_[-1],x2_[-1],Cord.loc[:,0],Cord.loc[:,1]))/\
            np.sum(C.loc[:,0]*D.loc[:,0]/euclidean_dist(x1_[-1],x2_[-1],Cord.loc[:,0],Cord.loc[:,1]))   ### new x1
        x2_new=np.sum(C.loc[:,0]*D.loc[:,0]*Cord.loc[:,1]/euclidean_dist(x1_[-1],x2_[-1],Cord.loc[:,0],Cord.loc[:,1]))/\
            np.sum(C.loc[:,0]*D.loc[:,0]/euclidean_dist(x1_[-1],x2_[-1],Cord.loc[:,0],Cord.loc[:,1]))   ### new x2
        
        x1_.append(x1_new)
        x2_.append(x2_new)
        
    return(pd.DataFrame({'x1':x1_,'x2':x2_}))
```

```python
Weiszfeld_Algo()
```
