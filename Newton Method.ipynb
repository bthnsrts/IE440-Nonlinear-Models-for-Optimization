```python
import math
import pandas as pd

def f(x):
    #the function given
    return (x**3)*math.cos(x)*math.sin(x) + 3*(x**2)*math.sin(x)-3*x

def df(x):
    ## the first derivative of the function 
    return (-1*x*(math.sin(x)*(x*(x*math.sin(x) - 3*math.cos(x)) -6) - x*math.cos(x)*(x*math.cos(x) + 3)) -3)

def ddf(x):
    ## the second derivative of the function
    return (-6*(x**2)*(math.sin(x)**2) + ((6*x-4*(x**3))*math.cos(x) -3*(x**2)+6) + 6*(x**2)*(math.cos(x)**2)+ 12*x*math.cos(x))
```

```python
def newton_method(f, df, ddf, x_prev=10, e=1e-12):
    # f is the function to apply the newton method
    # df is the first derivative function
    # ddf is the second derivative function
    # x_0 is the starting point for the newton method 
    # epsilon is used to know when to stop, it's value can determine the precision

    #create empty lists to report
    list_xk = [] ## starting x value at kth iteration
    list_f = [] ## f(x) values
    list_df = [] ## f'(x) values
    list_ddf = [] ## f''(x) values
    list_p = [] ## convergence ratio
    i = 0 ## iteration number
    x_k = x_prev ## for the first iteration there is no previous starting point
    last_iter = 0 ## 1 if conditions are met
    while(True):
        first = df(x_k) ## first derivative at the starting point
        second = ddf(x_k) ## second derivative at the starting point
        x_root = x_k - (first/second) ## next x is calculated
        list_xk.append(x_k) 
        y = f(x_k)
        list_f.append(y)
        list_df.append(first)
        list_ddf.append(second)
        if(i == 0):
            ## if first iteration, there is no convergence rate
             p = float("NAN") 
        else:
             p = abs(x_root - x_k)/pow(abs(x_k - x_prev),2) ## convergence rate
        list_p.append(p)
        
        if(last_iter == 1):
            break ## close the loop if this was the last iteration
            
        if(abs(first) <= e):
            last_iter = 1 ## move to the last iteration if first derivative is small enough
        elif(abs(f(x_root)-y) <= e):
            last_iter = 1 ## move to the last iteration if difference between function outputs is small enough
        elif(abs(x_root-x_k) <= e):
            last_iter = 1 ## move to the last iteration if difference between x values is small enough
        else:
            ## iterate one step next by moving to x
            flag = x_k
            x_k = x_root
            x_prev = flag
            i = i + 1
    return pd.DataFrame({'x_k':list_xk, 'f(x_k)':list_f, 
                         "f_'(x_k)": list_df, "f_''(x_k)": list_ddf,
                         "|(x_k+1-x_k)|/|(x_k-x_k-1)|^(2)": list_p}).rename_axis('Iteration')
```

```python
import matplotlib.pyplot as plt
from matplotlib.pyplot import text

def draw_graph(f, a=-10, b=10, x_k = float('inf'), f_name = "f(x)"):
    X = [a + 0.001*x for x in range(0,(b-a)*1000)]
    Y = [f(x) for x in X]
    fig = plt.figure(figsize = [10,10]) 
    plt.plot(X,Y, "b", linewidth = 2)
    
    plt.axvline(x=a, ymin=-2000, ymax=2000, linewidth=1, linestyle = ":", color = "k")
    plt.axvline(x=b, ymin=-2000, ymax=2000, linewidth=1, linestyle = ":", color = 'k')
    plt.axvline(x=0, ymin=-2000, ymax=2000, linewidth=1, color = 'k')
    plt.axhline(y=0, xmin=-15, xmax=15, linewidth=1, color = 'k')
    plt.axvline(x=x_k, ymin=-2000, ymax=2000, linewidth=1, linestyle = ":", color = 'r')
    
    font = {'color':'darkred','size':20}
    font2 = {'color':'darkred','size':15}
    plt.title("Graph of the function", fontdict = font)
    plt.xlabel("X axis", fontdict = font)
    plt.ylabel(f_name, fontdict = font) 
    text(x_k+0.1, 20, "x*", rotation=0, verticalalignment='center', fontdict = font2)
    plt.show()
```

```python
report = newton_method(f, df, ddf)
report
```

```python
def check_extremum(report, f):
    interval = [f(report.iloc[-1,0] - 1e-6), f(report.iloc[-1,0]), f(report.iloc[-1,0] + 1e-6)]
    
    if (interval[0] < interval[1]) and (interval[2] < interval[1]):
        return ("Solution is local max!!!!")
    elif (interval[0] > interval[1]) and (interval[2] > interval[1]):
         return ("Solution is local min!!!!")
    else:
        return ("Solution is a saddle point!!!")
```

```python
def main():
    ## 1. Draw the graph of the function between values a = -10, b = 10
    draw_graph(f,-10,10)
    ## 2. Determine at least three different parameter sets for Secant method.
    ## Each list below contains x_0, e respectively
    parameters =[[10,1e-12], [5,1e-12], [-5,1e-12]]
    for params in parameters:
        report = newton_method(f, df, ddf, params[0], params[1])
        print(f" e = {params[1]}, x_0 = {params[0]}")
        display(report)
        print(f" x* = {report.iloc[-1, 0]}, f(x) = {report.iloc[-1, 1]}")
        print(check_extremum(report,f))
        draw_graph(f, -10, 10, x_k = report.iloc[-1, 0])
        draw_graph(df, -10, 10, x_k = report.iloc[-1, 0], f_name = "f'(x)")
```
