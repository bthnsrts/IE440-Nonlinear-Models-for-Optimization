```python
import math
import pandas as pd

def function(x):
    return (x**3)*math.cos(x)*math.sin(x) + 3*(x**2)*math.sin(x)-3*x

def first_derivative(x):
    ## the first derivative of the function 
    return (-1*x*(math.sin(x)*(x*(x*math.sin(x) - 3*math.cos(x)) -6) - x*math.cos(x)*(x*math.cos(x) + 3)) -3)

def second_derivative(x):
    ## the second derivative of the function
    return (-6*(x**2)*(math.sin(x)**2) + ((6*x-4*(x**3))*math.cos(x) -3*(x**2)+6) + 6*(x**2)*(math.cos(x)**2)+ 12*x*math.cos(x))
```

```python
def newton_method(func, first_derivative, second_derivative, x_0=10, e=1e-12):
    iter_ = [] ## iterations list
    x_0l = []
    x_1l = []
    x_kl = [] ## x_k values for k = 1 to the last iteration
    f = [] ## f(x_k) values
    first_derivatives = [] ## f'(x_k) values
    second_derivatives = [] ## f''(x_k) values
    r = [] ## (x_k+1 - x_k) / (x_k - x_k-1)^2 ratio
    i = 0 ## iteration number
    x_1 = x_0
    while(True):
        x = x_1 - first_derivative(x_1)/second_derivative(x_1)
        iter_.append(i)
        x_kl.append(x) ## x_k
        if(i == 0):
            x_0l.append(float("NAN"))
        else:
            x_0l.append(x_0)
        x_1l.append(x_1)
        f.append(func(x))
        first_derivatives.append(first_derivative(x_1))
        second_derivatives.append(second_derivative(x_1))
        if(i == 0):
            r.append(float("NAN"))
        else:
            r.append(abs(x-x_1)/pow(abs(x_1-x_0),2))
        
        if(abs(first_derivative(x)) <= e):
            break
        elif(abs(func(x)-func(x_1)) <= e):
            break
        elif(abs(x-x_1) <= e):
            break
        else:
            flag = x_1
            x_1 = x
            x_0 = flag
            i = i + 1
    return pd.DataFrame({"iteration":iter_, 'x_k-2': x_0l, 'x_k-1':x_1l, 'x_k':x_kl, 'f(x_k)':f, 
                         "f_derivative(x_k)": first_derivatives, "f_second_derivative(x_k)": second_derivatives,
                         '(x_k+1-x_k)/(x_k-x_k-1)^(2)': r})
```

```python
import matplotlib.pyplot as plt
from matplotlib.patches import Polygon
from matplotlib.pyplot import text

def draw_graph(func, a=-10, b=10, x_k = float('inf'), f = "f(x)"):
    X = [a + 0.001*x for x in range(0,(b-a)*1000)]
    Y = [func(x) for x in X]
    fig = plt.figure(figsize = [10,10]) 
    plt.plot(X,Y, "b", linewidth = 2)
    
    plt.axvline(x=a, ymin=-2000, ymax=2000, linewidth=1, linestyle = ":", color = "k")
    plt.axvline(x=b, ymin=-2000, ymax=2000, linewidth=1, linestyle = ":", color = 'k')
    plt.axvline(x=0, ymin=-2000, ymax=2000, linewidth=1, color = 'k')
    plt.axhline(y=0, xmin=-15, xmax=15, linewidth=1, color = 'k')
    plt.axvline(x=x_k, ymin=-2000, ymax=2000, linewidth=1, linestyle = ":", color = 'r')
    
    font = {'color':'darkred','size':20}
    font2 = {'color':'darkred','size':15}
    plt.title("Graph of the function", fontdict = font)
    plt.xlabel("X axis", fontdict = font)
    plt.ylabel(f, fontdict = font) 
    text(x_k+0.1, 20, "x*", rotation=0, verticalalignment='center', fontdict = font2)
    plt.show()
```

```python
report = newton_method(function, first_derivative, second_derivative)
report
```

```python
def main():
    ## 1. Draw the graph of the function between values a = -10, b = 10
    draw_graph(function,-10,10)
    ## 2. Determine at least three different parameter sets for Secant method.
    ## Each list below contains e, x(0) respectively
    parameters =[[1e-12,10], [1e-12,-5], [1e-12,5]]
    for params in parameters:
        report = newton_method(function, first_derivative, second_derivative, params[1], params[0])
        print(f" e = {params[0]}, x_0 = {params[1]}")
        display(report)
        print(f" x* = {report.iloc[-1, 3]}, f(x) = {function(report.iloc[-1, 3])}")
        draw_graph(function, -10, 10, x_k = report.iloc[-1, 1])
        draw_graph(first_derivative, -10, 10, x_k = report.iloc[-1, 3], f = "f'(x)")
```

```python
main()
```
