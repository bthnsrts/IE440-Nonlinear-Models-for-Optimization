```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

#the function
def f(x):
    return np.power(x,3)*np.cos(x)*np.sin(x)+3*np.power(x,2)*np.sin(x)-3*x

#first derivative of f
def df(x):
    return -np.power(x,3)*np.power(np.sin(x),2)+(3*np.power(x,2)*np.cos(x)+6*x)*np.sin(x)+\
           np.power(x,3)*np.power(np.cos(x),2)+3*np.power(x,2)*np.cos(x)-3




```

```python
def Golden_Section(f, a, b, epsilon):
    gamma = (1 + (5**(0.5)))/2  
    c = 1/gamma
    
    x = b - c*(b-a) # defining the x_0
    y = a + c*(b-a) # defining the y_0
    
    fx = f(x)       # defining the f(x_0)
    fy = f(y)       # defining the f(y_0)
    
    num_of_iterations = 0 # keeps the info of # of iterations.
    
    a_s = [a]  # These lists are constructed to hold the informations regarding the iterations.
    b_s = [b]
    x_s = [x]
    y_s = [y]
    fx_s = [fx]
    fy_s = [fy]
    
    
    while b-a > epsilon: 
        num_of_iterations += 1
        
        if fx > fy: 
            
            a = x   # moving the point a to the x
            x = y   # moving the x to the y, will be used for the next iteration
            y = a + c*(b-a) # defining the new y that will be used for the next iteration
            
            fx = fy # defining the fx that will be used for the next iteration. We do not have to compute the new fx
                    # by putting into the f(x) because new x is the old y and new fx is already equals to old fy. 
            fy = f(y) # new y point is determined so we need to compute that. Will be used for the next iteration
            
        elif fx <= fy:
            
            b = y # moving the point b to the y
            y = x # moving the y to the x, will be used for the next iteration
            x = b - c*(b-a) # defining the new x that will be used for the next iteration
            
            fy = fx # defining the fy that will be used for the next iteration. We do not have to compute the new fy
                    # by putting into the f(y) because new y is the old x and new fy is already equals to old fx. 
            fx = f(x)# new x point is determined so we need to compute that. Will be used for the next iteration
            
        a_s.append(a)  # adding the iteration information to the lists
        b_s.append(b) 
        x_s.append(x) 
        y_s.append(y) 
        fx_s.append(fx)
        fy_s.append(fy) 
            
    x_star = x # iterations are completed. x_star is the local optima point 
    f_x_star = f(x_star) # local optima value
    
    #Creating a pandas dataframe to output the desired format
    iterations_df = pd.DataFrame({"a": a_s, "b": b_s, "x": x_s, "y": y_s, "f_x": fx_s, "f_y": fy_s, 
                                 "|(x_k+1 - x_k)| / |(x_k - x_k-1)|": np.nan, "-log|x_k+1 - x_k| + log|x_k - x_k-1|": np.nan})
    
    
    for i in range(1,len(iterations_df)-1): # iterating over every row of the dataframe to compute
                                            #"(x_k+1 - x_k) / (x_k - x_k-1)" and "-log|x_k+1 - x_k| + log|x_k - x_k-1|"
                                              #please realize i starts from 1 not 0 and i goes upto the 
                                              # len(iterations_df)-1 in order to be able to compute x_k-1 and x_k+1
                
        x_k_plus_1  = iterations_df.iloc[i+1, 2] # i + 1 represents the row and 2 is the index of column "x"
        x_k         = iterations_df.iloc[i, 2]   # i represents the row and 2 is the index of column "x"
        x_k_minus_1 = iterations_df.iloc[i-1, 2] # i - 1 represents the row and 2 is the index of column "x"
    
        iterations_df.iloc[i, 6] = abs(x_k_plus_1 - x_k) / abs(x_k - x_k_minus_1) # i represents the row and 6 is the 
                                                                        #index of column "|(x_k+1 - x_k)| / |(x_k - x_k-1)|"
        
        iterations_df.iloc[i, 7] = (-1*np.log(abs(x_k_plus_1 - x_k))) + (np.log(abs(x_k - x_k_minus_1))) 
                                                                        # i represents the 
                                                                        #row and 5 is the index of column
                                                                        # "-log|x_k+1 - x_k| + log|x_k - x_k-1|"

    
    print(iterations_df)
    print("x* = ", x_star)
    print("f(x*) = ", f_x_star)
    
    
    return {"x*": x_star, "f(x*)": f_x_star}
            
            
            
        
        
    
```

```python
output = Golden_Section(f, -10, +10, 0.001)
```

```python
a=-10
b=10
root=output["x*"]
num_pts = int(1e4)
x = [a+(b-a)*i/num_pts for i in range(num_pts)]
y_f = [f(xn) for xn in x] 
y_df = [df(xn) for xn in x]


fig, axes = plt.subplots(1, 2, figsize=(15,6))

axes[0].axhline(0, color='black')
axes[0].axvline(root, color='g', label='local minimum')
axes[0].plot(x,y_f)
axes[0].set_title('Graph of f')
axes[0].set_xlabel('x axis')
axes[0].set_ylabel('f')
axes[0].legend(loc='lower right')

axes[1].axhline(0, color='black')
axes[1].axvline(root, color='g', label='root')
axes[1].plot(x,y_df)
axes[1].set_title('Graph of the derivative of f')
axes[1].set_xlabel('x axis')
axes[1].set_ylabel('derivative of f')
axes[1].legend(loc='lower right')

plt.show()
```
